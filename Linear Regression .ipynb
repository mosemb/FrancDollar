```python
X= df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]

y= df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Split the data into training and testing sizes
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=109)
```

#Now we import the linear regression model and put it in variable 

```python
from sklearn.linear_model import LinearRegression
```

```python
linearmodel = LinearRegression()
```

```python
linearmodel.fit(X_train,y_train)
```

```python
print(linearmodel.intercept_)
```

```python
linearmodel.coef_
```

```python
X_train.columns
```

```python
import pandas as pd
coefdataframe = pd.DataFrame(linearmodel.coef_,X.columns,columns=['Coeff'])
```

```python
coefdataframe
```

# Predictions  for the dataset , predicting prices using X_test

```python
prediction = linearmodel.predict(X_test)
prediction 
```

# Visualise the prediction with the real data on a scatter plot 

```python
plt.scatter(y_test,prediction)
plt.show()
```

# Its as straight line between the ytest Vs prediction, so we are doing good. 

# The residuals or error between the y_test and prediction

```python
error = (y_test -prediction)
```

```python
sns.distplot(error)
```

# Note that the residuals are normally distributed so that means that the model that we chose is the right choice for the problem at hand. If they would have been otherwise then we would try another model. 

## Now we calculate the Mean Absolute Error MAE, Mean Squared Error MSE, Root Mean Squared Error RMSE. 

```python
from sklearn import metrics 
```

```python
metrics.mean_absolute_error(y_test,prediction)
```

```python
mse = metrics.mean_squared_error(y_test,prediction)
mse
```

```python
np.sqrt(mse)
```
