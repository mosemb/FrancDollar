```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age  
    
```

```python
df['Age']= df[['Age','Pclass']].apply(impute_age,axis=1)
```

## There are a few issues with the daframe. One there are columns in the frame that are not useful for the machine learning algorithm right now,  2). There are columns that we need to convert to numerical values because the machine learning algorithm will not be able to read it in that state. These are Embarked and Sex. We will use get_dummies for this task. 


```python
pd.get_dummies(df['Sex'])
```

## So male and female have been converted to either 1 or 0 depending on the column used. So since if 1 is on its automatic 0 is the corresponding column in the other column. 

## So we can drop one column since either one conveys the message perfectly to us. 

```python
sex = pd.get_dummies(df['Sex'],drop_first=True)
```

```python
pd.get_dummies(df['Embarked'])
```

```python
embark = pd.get_dummies(df['Embarked'],drop_first=True)
```

## Now we concatnated embark with sex as the new data frames to train. 

```python
df = pd.concat([sex,embark,df],axis=1)
```

```python
df.head()
```

## Now we can drop the other columns that are not going to be read by the algorithm that is Name, Sex , Ticket , Embarked


```python
df.drop(['Name','Sex','Ticket','Embarked'], axis=1,inplace=True)
```

```python
df.head()
```

## At this point still lets drop the column that has the PassengerId because it simply stores the ids which we can find in the dataframe as well.
